```python
# import pandas as pd
#NOTA: antes de ejecutar el script, copiar la imagen en clipboard.(click derecho copiar, o ctrl+c)
#nombre del excel que contiene los nombres de los grupos de whatsapp
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
import re
from selenium.common.exceptions import NoSuchElementException
import time
linkwsp = []
namegroup = []
cantCreateGroups=20
df = pd.read_excel('grupowsp.xlsx')
nameCollaboratorWsp = "ingGianmarco Arias"
###########################################CREACION DE METODOS#################################################################
def deleteGroup():
    #da click en el nombre del grupo
    driver.find_element(by=By.XPATH, value='//*[@id="main"]/header/div[2]/div[1]/div/span').click()
    time.sleep(1.000)
    #Dar click salir grupo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[7]/div[1]/div[2]/div/span').click()
    time.sleep(0.8)
    #confimar salir grupo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[3]/div/div[3]/div/div/div').click()
    time.sleep(1.8)
    #Eliminar grupo
    try:
        driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[5]/div[1]/div[2]/div/span').click()
        time.sleep(1.5)
    except:
        driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[1]/div[2]/div/span').click()
        time.sleep(1.5)
    #Confirmar la eliminacion del grupo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[3]/div/div[2]/div/div').click()
    time.sleep(2.000)
    
def createGroup(namegroups):
    print("CreateGroup")
    #Click crear chat
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[3]/header/div[2]/div/span/div[3]/div/span').click()
    time.sleep(0.5)
    #click crear Grupo    
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[1]/span/div/span/div/div[2]/div[1]/div/div[1]/div[2]/div/div').click()
    time.sleep(0.5)
    #Escribir nombre del colaborador para el grupo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[1]/span/div/span/div/div/div[1]/div/div/input').send_keys(nameCollaboratorWsp)
    time.sleep(1)
    #Dar click en el unico resultado
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[1]/span/div/span/div/div/div[2]/div/div/div/div/div/div/div[2]/div[1]/div/span').click()
    time.sleep(1)
    #dar boton siguiente
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[1]/span/div/span/div/div/span/div/span').click()
    time.sleep(1)
    #escribir nombre de grupo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[1]/span/div/span/div/div/div[1]/div[2]/div/div[2]/div[1]/div/div[2]').send_keys(str(namegroups))
    time.sleep(1)
    #click confirmar creacion de gruo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[1]/span/div/span/div/div/span/div/div/span').click()
    time.sleep(2.5)
    # click en el nombre del grupo
    driver.find_element(by=By.XPATH, value='//*[@id="main"]/header/div[2]/div[1]/div/span').click()
    time.sleep(1.0)
    #Click Enlace de invitacion
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[2]/div[2]/div/div').click()
    time.sleep(2)
    #obtener link del enlace
    getLinkWSP = driver.find_element(by=By.XPATH, value='//*[@id="group-invite-link-anchor"]')
    
    linkwsp.append(getLinkWSP.text)
    namegroup.append(namegroups)
    
    #Click retroceder 
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/header/div/div[1]/button/span').click()
    try:
        for j in range(int(11)):
            print("entra al primer for "+str(j+1))
            #mouse hover
            time.sleep(2)
            ele =driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]')    
            ActionChains(driver).move_to_element(ele).perform()
            time.sleep(1)
            #abrir lista de acciones a realizar al usuario
            try:
                driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]/div[2]/div[2]/span[2]').click()
            except:
                driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]/div[2]/div[2]/span[3]').click()
            time.sleep(0.800)

            #da click en eliminar
            ll = driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[2]/div[1]')
            if ll.text=="Descartar como admin.":
                driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[1]/div[1]').click()
            else:
                driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[2]/div[1]').click()
            time.sleep(1.800)
            #da click en confirmar la eliminacion
            driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[2]/div/div[2]/div/div').click()
            time.sleep(1.300)
    except:
        print('')
def addLinkExcel():
    df = pd.DataFrame({ 'NAME_GROUP_WSP': namegroup,'LINK_GROUP_wSP': linkwsp})
    df = df[['NAME_GROUP_WSP', 'LINK_GROUP_wSP']]
    writer = pd.ExcelWriter('EXCEL_GENERADOS/LINKWSP'+time.strftime("%Y%m%d-%H-%M-%S")+'.xlsx')
    df.to_excel(writer, 'Sheet', index=False)
    #writer.save()
    writer.close()
###########################################CREACION DE METODOS#################################################################
try:
    for row, datos in df.iterrows():
        #nombre de la columna
        namewsp = datos['NOMBRE_DE_GRUPO']
        time.sleep(2)
        #da click del primer resultaldo, si son varios resultados por lo general el resultado es del maximo hacia arriba
        #si hay 6 resultados, la fila del resultado a dar clic seria el 6
        if(cantCreateGroups > row):
            createGroup(namewsp)
            time.sleep(0.8)
        else:    
        #####AL TERMINAR DE AGREGAR TODOS LOS LINK CREAMOS NUESTRO EXCEL CON LOS ARRAYS OBTENIDOS
            print('')
    addLinkExcel()
except:
    addLinkExcel()
```
