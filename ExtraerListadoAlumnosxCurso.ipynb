```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import os
simp_path = '.env'
abs_path = os.path.abspath(simp_path)
CREDENTIAL={'EMAIL':'','PASSWORD':''}
with open(abs_path,"r") as archivo:
    for linea in archivo:        
        if (str(linea.strip('\n')).split('='))[0] == "EMAIL":
            CREDENTIAL["EMAIL"]=(str(linea.strip('\n')).split('='))[1]
        if (str(linea.strip('\n')).split('='))[0] == "PASSWORD":
            CREDENTIAL["PASSWORD"]=(str(linea.strip('\n')).split('='))[1]
            
            
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
#abrir formulario

driver.get('https://sivireno.undc.edu.pe/index_login.php')
driver.find_element(by=By.XPATH, value='/html/body/div[1]/div/div/form/div[1]/a').click()
driver.find_element(by=By.XPATH, value='//*[@id="identifierId"]').send_keys(CREDENTIAL["EMAIL"])
driver.find_element(by=By.XPATH, value='//*[@id="identifierNext"]/div/button/span').click()
for j in range(10):
    try:
        driver.find_element(by=By.XPATH, value='//*[@id="password"]/div[1]/div/div[1]/input').send_keys(CREDENTIAL["PASSWORD"])
        driver.find_element(by=By.XPATH, value='//*[@id="passwordNext"]/div/button/span').click()
        break
    except:
        time.sleep(1)
```

```python
import pandas as pd
import pdfkit
import base64
import os
from bs4 import BeautifulSoup
from selenium import webdriver

# Configuración de pdfkit
path_to_wkhtmltopdf = r'C:\Program Files\wkhtmltopdf\bin\wkhtmltopdf.exe'  # Cambia esta ruta si es necesario
config = pdfkit.configuration(wkhtmltopdf=path_to_wkhtmltopdf)

options = {
    'enable-local-file-access': None,
    'no-stop-slow-scripts': None,
}

# Función para convertir imágenes a Base64
def image_to_base64(img_path):
    with open(img_path, "rb") as img_file:
        return base64.b64encode(img_file.read()).decode('utf-8')

# Cargar el CSS
def load_css(css_path):
    with open(css_path, 'r', encoding='utf-8') as css_file:
        return css_file.read()

# Cargar datos
df = pd.read_excel('EXCEL_TRABAJADO/2024-2.xlsx')

# Ruta a tu CSS y a la imagen
css_path = 'style/css1.css'  # Cambia la ruta según sea necesario
logo_path = 'style/logo_undc.png'  # Cambia la ruta según sea necesario

# Obtener el CSS
css_content = load_css(css_path)

# Obtener la imagen en formato Base64
logo_base64 = image_to_base64(logo_path)

# Iterar sobre los datos
for row, datos in df.iterrows():
    if datos['FLAG'] == 'SI':
        try:
            url = datos['LIST_STUDENTS']
            driver.get(url)

            # Obtener el HTML completo de la página
            html_content = driver.page_source
            
            # Usar BeautifulSoup para procesar el HTML
            soup = BeautifulSoup(html_content, 'html.parser')

            # Seleccionar solo la segunda, tercera, cuarta y quinta tablas
            tables = soup.find_all('table')
            # Añadimos un salto de línea entre cada tabla (<br>)
            filtered_html = '<br>'.join(str(tables[i]) for i in range(1, 5))

            # Construir el HTML final con CSS e imagen
            full_html_content = f"""
            <!DOCTYPE html>
            <html lang="es">
            <head>
                <meta charset="UTF-8">
                <title>Registro de Asistencia</title>
                <style>
                    {css_content}
                </style>
            </head>
            <body>
                <table width="100%">
                    <tr>
                        <td align="left" width="20%">
                            <img src="data:image/png;base64,{logo_base64}" alt="Logo" width="50" height="50">
                        </td>
                        <td align="center" width="60%">
                            <h2>UNIVERSIDAD NACIONAL DE CAÑETE</h2>
                        </td>
                        <td align="right" width="20%">
                            <strong>ORA<br></strong>
                            <strong>Fecha:</strong> 23/09/2024 4:31pm<br>
                            <strong>Usuario:</strong> 73623097
                        </td>
                    </tr>
                </table>
                {filtered_html} <!-- Aquí se agrega solo el contenido de las tablas filtradas -->
            </body>
            </html>
            """

            # Guardar el HTML como PDF
            output_pdf = f"PDF_GENERADOS/{datos['CREACION_NOMBRE_DEL_GRUPO']}.pdf"
            pdfkit.from_string(full_html_content, output_pdf, configuration=config, options=options)
            print(f"Guardado: {output_pdf}")
        except Exception as e:
            print(f"Error al guardar PDF: {e}")

# Cerrar el navegador
driver.quit()
```
