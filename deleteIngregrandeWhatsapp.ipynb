```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
driver = webdriver.Chrome(executable_path='chromedriver.exe')
#abrir formulario
time.sleep(2.5)
url="https://web.whatsapp.com/"
driver.get(url)
```

```python
##SCRIPT PARA ELIMINAR PARTICIPANTES DE UN GRUPO

# import pandas as pd
#NOTA: antes de ejecutar el script, copiar la imagen en clipboard.(click derecho copiar, o ctrl+c)
#nombre del excel que contiene los nombres de los grupos de whatsapp
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
import re
from selenium.common.exceptions import NoSuchElementException
df = pd.read_excel('grupowsp.xlsx')
#nombre de la columna
    
characters = "'!?() "
def deleteIntegrante():
#######################################################
    try:
        dd=driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/button/div[2]')
        cant= re.sub(r'[a-zA-Z-รกรบ]','',dd.text).rstrip()
        for x in range(len(characters)):
            cant = cant.replace(characters[x],'')
        cant = int(cant)
        cant = cant-1
        if int(cant) >10:
            for j in range(int(cant)):
                print("entra al primer for "+str(j+1))
                #mouse hover
                ele =driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]')    
                ActionChains(driver).move_to_element(ele).perform()
                time.sleep(0.500)
                #abrir lista de acciones a realizar al usuario
                try:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]/div[2]/div[2]/span[2]').click()
                except:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]/div[2]/div[2]/span[3]').click()
                time.sleep(0.400)
                #da click en eliminar
                ll=driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[1]/div[1]')
                if ll.text=="Descartar como admin.":
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[1]/div[1]').click()
                else:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[2]/div[1]').click()
                time.sleep(0.800)
                #da click en confirmar la eliminacion
                driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[2]/div/div[2]/div/div').click()
                time.sleep(1.000)
            
        for j in range(11):
            if (11-j) != 10:
                print("entra al segundo for "+str(11-j))
                #mouse hover
                ele =driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div['+str(11-j)+']/div/div/div[2]')    
                ActionChains(driver).move_to_element(ele).perform()
                time.sleep(0.5)
                #abrir lista de acciones a realizar al usuario
                try:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div['+str(11-j)+']/div/div/div[2]/div[2]/div[2]/span[2]').click()
                except:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div['+str(11-j)+']/div/div/div[2]/div[2]/div[2]/span[3]').click()
                time.sleep(0.400)
                #da click en eliminar
                ll = driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[2]/div[1]')
                if ll.text=="Descartar como admin.":
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[1]/div[1]').click()
                else:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[2]/div[1]').click()
                time.sleep(0.800)
                #da click en confirmar la eliminacion
                driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[2]/div/div[2]/div/div').click()
                time.sleep(1.000)
    #Cuando es menor <= 11
    except NoSuchElementException:
        try:
            for j in range(int(11)):
                print("entra al tercer for "+str(j+1))
                #mouse hover
                ele =driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]')    
                ActionChains(driver).move_to_element(ele).perform()
                time.sleep(0.500)
                #abrir lista de acciones a realizar al usuario
                try:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]/div[2]/div[2]/span[2]').click()
                except:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[2]/div[3]/span/div/span/div/div/section/div[6]/div[2]/div[3]/div/div[1]/div/div/div[2]/div[2]/div[2]/span[3]').click()
                time.sleep(0.200)

                #da click en eliminar
                ll = driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[2]/div[1]')
                if ll.text=="Descartar como admin.":
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[1]/div[1]').click()
                else:
                    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[4]/div/ul/div/li[2]/div[1]').click()
                time.sleep(0.800)
                #da click en confirmar la eliminacion
                driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[2]/div/div[2]/div/div').click()
                time.sleep(1.000)
        except:
            print('Entra aqui except tercer')

    #######################################################

for row, datos in df.iterrows():
    namewsp = datos['NOMBRE_DE_GRUPO']
    #buscador de Chat (persona/grupo)
    driver.find_element(by=By.XPATH, value='//*[@id="side"]/div[1]/div/div/div[2]/div/div[2]').clear()
    time.sleep(1)
    driver.find_element(by=By.XPATH, value='//*[@id="side"]/div[1]/div/div/div[2]/div/div[2]').send_keys(namewsp)
    time.sleep(1)
    #da click del primer resultaldo
    for j in range(int(11)):
        try:
            #8 RESULTADO ADICIONAL
            driver.find_element(by=By.XPATH, value='//*[@id="pane-side"]/div[1]/div/div/div['+str(11-j)+']').click()
            time.sleep(1)
            #da click en el nombre del grupo
            driver.find_element(by=By.XPATH, value='//*[@id="main"]/header/div[2]/div[1]/div/span').click()
            time.sleep(1)
            deleteIntegrante()
        except:
            print("")
        
```
