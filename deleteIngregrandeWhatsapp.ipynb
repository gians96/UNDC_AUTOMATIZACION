```python
##ELIMINAR GRUPO

# import pandas as pd
#NOTA: antes de ejecutar el script, copiar la imagen en clipboard.(click derecho copiar, o ctrl+c)
#nombre del excel que contiene los nombres de los grupos de whatsapp
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
import re
from selenium.common.exceptions import NoSuchElementException
df = pd.read_excel('grupowsp.xlsx')
#nombre de la columna

def deleteGroup():
    time.sleep(1.500)
    #Dar click salir grupo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[5]/span/div/span/div/div/section/div[7]/div[1]/div[2]/div/span').click()
    time.sleep(1.2)
    #confimar salir grupo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[3]/div/div[3]/div/div/div').click()
    time.sleep(1.3)
    #Eliminar grupo
    try:
        driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[5]/span/div/span/div/div/section/div[5]/div[1]/div[2]/div/span').click()
        time.sleep(1.5)
    except:
        driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div/div[5]/span/div/span/div/div/section/div[6]/div[1]/div[2]/div/span').click()
        time.sleep(1.5)
    #Confirmar la eliminacion del grupo
    driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/span[2]/div/div/div/div/div/div/div[3]/div/div[2]/div/div').click()
    time.sleep(2.000)
for row, datos in df.iterrows():
    namewsp = datos['NOMBRE_DE_GRUPO']
    time.sleep(2)
    #buscador de Chat (persona/grupo)
    driver.find_element(by=By.XPATH, value='//*[@id="side"]/div[1]/div/div/div[2]/div/div[2]').clear()
    time.sleep(1)
    driver.find_element(by=By.XPATH, value='//*[@id="side"]/div[1]/div/div/div[2]/div/div[2]').send_keys(namewsp)
    time.sleep(0.5)
    for j in range(int(11)):
        try:
            #8 RESULTADO ADICIONAL
            #da click en el nombre del grupo
            if driver.find_element(by=By.XPATH, value='//*[@id="pane-side"]/div[1]/div/div/div['+str(11-j)+']/div/div/div/div[2]/div[1]/div[1]/span/span/span').text == namewsp:
                driver.find_element(by=By.XPATH, value='//*[@id="pane-side"]/div[1]/div/div/div['+str(11-j)+']').click()
                time.sleep(1)
                driver.find_element(by=By.XPATH, value='//*[@id="main"]/header/div[2]/div[1]/div/span').click()
                time.sleep(1)
                deleteGroup()
        except NoSuchElementException:
            print('siguiente')                      
    time.sleep(0.8)
```
